```python
import pandas as pd
```

```python
compound_embedding_file = '/datasets/BBBC021/metadata/BBBC021_v1_compound_latent_rdkit.csv'
in_dist_compound_list = ['az138', 'az841', 'cytochalasin-b', 'demecolcine', 'epothilone-b', 'nocodazole', 'taxol', 'az258', 'colchicine', 'cytochalasin-d', 'docetaxel', 'latrunculin-b', 'pp-2', 'vincristine']
new_df = pd.read_csv(compound_embedding_file)
new_df = new_df.set_index('compound')
new_df
```

```python
# calculate the correlation of compound embedding with each other
new_df = new_df.T
correlation_matrix = new_df.corr()
correlation_matrix
```

```python
# only select rows from in_dist_compound_list
correlation_matrix = correlation_matrix.loc[in_dist_compound_list]
correlation_matrix = correlation_matrix.drop(in_dist_compound_list, axis=1)
correlation_matrix
```

```python
# create a new dataframe with columns 'compound_1', 'compound_2', 'correlation' and create that from correlation_matrix
correlation_list = [(row, col, correlation_matrix.loc[row, col]) for row in correlation_matrix.index for col in correlation_matrix.columns if not pd.isnull(correlation_matrix.loc[row, col])]
correlation_list

```

```python
# sort the correlation_list based on correlation value and print the top 5 values
correlation_list.sort(key=lambda x: x[2], reverse=True)
correlation_list
```

```python
# wrtie correlation_list into a dataframe
correlation_df = pd.DataFrame(correlation_list, columns=['compound_1', 'compound_2', 'correlation'])
correlation_df
```

```python
top_15_ood = []
new_top_15_ood = pd.DataFrame(columns=['compound_1', 'compound_2', 'correlation'])
# loop over rows of the correlation_df dataframe and add the compound_2 to top_15_ood if it is not in in_dist_compound_list
for index, row in correlation_df.iterrows():
    if row['compound_2'] not in top_15_ood:
        top_15_ood.append(row['compound_2'])
        new_top_15_ood.loc[new_top_15_ood.shape[0]] = row
    if len(top_15_ood) == 15:
        break
    
new_top_15_ood
```
